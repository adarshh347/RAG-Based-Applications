```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.chains import RetrievalQA
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFaceHub
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()


def setup_qa_system(file_path):
    # Step 1: Load and split the document
    loader = PyPDFLoader(file_path)
    docs = loader.load_and_split()

    # Step 2: Split the text into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    chunks = text_splitter.split_documents(docs)

    # Step 3: Create embeddings
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")  # Free embedding model
    vector_store = FAISS.from_documents(chunks, embeddings)

    # Step 4: Set up the retriever and free Llama-like API model
    retriever = vector_store.as_retriever()

    # Use a free Llama-like model via HuggingFaceHub integration
    llm = HuggingFaceHub(
        repo_id="tiiuae/falcon-7b-instruct",  # Free instruct-tuned model
        model_kwargs={"temperature": 0.7, "max_length": 2048},
    )

    # Step 5: Set up the QA chain
    qa_chain = RetrievalQA.from_chain_type(llm, retriever=retriever)
    return qa_chain


if __name__ == "__main__":
    # Upload and save the file
    from google.colab import files

    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]

    # Save the file content to disk
    with open(file_name, "wb") as f:
        f.write(uploaded[file_name])

    # Set up the QA system
    qa_chain = setup_qa_system(file_name)

    # Interactive loop for QA
    while True:
        question = input("\nAsk a question: ")
        if question.lower() in ["exit", "quit"]:
            print("Exiting the QA system.")
            break

        # Run the QA chain
        try:
            answer = qa_chain.run(question)
            print("\nAnswer:")
            print(answer)
        except Exception as e:
            print(f"\nAn error occurred: {e}")
```
